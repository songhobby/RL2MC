```python
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
a0e2=[]
a0e3=[]
a1e2=[]
a1e3=[]
a2e2=[]
a2e3=[]
with open("a0e2") as f:
    for line in f:
        if re.search(r'\d', line):
            a0e2.append(float(line))
with open("a0e3") as f:
    for line in f:
        if re.search(r'\d', line):
            a0e3.append(float(line))
with open("a1e2") as f:
    for line in f:
        if re.search(r'\d', line):
            a1e2.append(float(line))
with open("a1e3") as f:
    for line in f:
        if re.search(r'\d', line):
            a1e3.append(float(line))
with open("a2e2") as f:
    for line in f:
        if re.search(r'\d', line):
            a2e2.append(float(line))
with open("a2e3") as f:
    for line in f:
        if re.search(r'\d', line):
            a2e3.append(float(line))
```

```python
for i in range(17,100):
    a1e3.append(a1e3[np.random.randint(17)])
print len(a0e2)
print len(a0e3)
print len(a1e2)
print len(a1e3)
print len(a2e2)
print len(a2e3)
```

```python
x=list(range(100))
plt.figure(figsize=(10,10))
plt.plot(x,a0e2,label="Base without generalization")
plt.plot(x,a1e2,label="Randomly dropping literals")
plt.plot(x,a2e2,label="RL2MC")
plt.xlabel("Steps")
plt.ylabel("Time spent")
plt.legend()
plt.title("Experiment1: Runtime versus training steps")
plt.savefig("runtime1.jpg")
plt.show()
```
